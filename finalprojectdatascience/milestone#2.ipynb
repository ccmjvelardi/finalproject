### Run the info function to check for null values

```python
data_df.info()
```

<br>
<br>
Since the overall goal of this project is to check for overall 5-Star Ratings, the rows without a rating will be dropped. With 4670 non-null values remaining, this will still be a good sample size to get reliable data.

```python
data_df = data_df.dropna(subset=['OVERALL_STARS'])
data_df.info()
```

<br>
<br>
Next will be filling in the non-null values accordingly. Going over thesee columns, filling them in as Not Available would make the most sense. 

```python
data_df = data_df.fillna('Not Available')
data_df.info()
```
